```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
depth = -0.0003048 * np.array([547, 817, 917, 1027, 1117, 1187, 1365, 1457])
temp = np.array([12.272, 18.066, 19.992, 21.827, 24.104, 25.485, 29.411, 31.039])
```

```python
sns.lineplot(y = depth, x = temp)
plt.show()
```

```python
x = depth.reshape((-1, 1))
```

```python
model = LinearRegression()
model.fit(x, temp)
```

```python
print(f"slope: {model.coef_}")
```
